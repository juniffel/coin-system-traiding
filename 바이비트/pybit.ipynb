```python
from pybit.unified_trading import HTTP, WebSocket
import pandas as pd
pd.set_option("display.max_columns", None)
# pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
########  Defines ###########################
testnet = False

###############################################
with open('key.txt') as f:
    lines = f.readlines()
    api_key = lines[0].strip()
    secret = lines[1].strip()
    
session = HTTP(testnet = testnet, api_key = api_key,api_secret = secret, logging_level = 10,)

```

```python
float(session.get_server_time()['time'])/1000
```

```python
import datetime as dt
dt.datetime.fromtimestamp(float(session.get_server_time()['time'])/1000), dt.datetime.now()
# session.get_server_time()
```

```python
def klines(category, symbol, interval, limit=1000 ):
	df = pd.DataFrame(session.get_kline(
	category="linear",
	symbol="BTCUSDT",
	interval=60,   
	)['result']['list'])
	# 열에 이름 지정 time.isoformat(1715663257.805)변환하고 ohlcv가 모두 숫자인지 확인합니다. 
	df['date'] = pd.to_datetime(df['date'], unit='ms') 
	for col in df.columns[1:]: 
		df[col ] = pd.to_numeric(df[col])
	return df

klines(category="linear",
	symbol="BTCUSDT",
	interval=60,
 	limit=10)
```
